### Координаты слов на плоскости (2-я и 3-я компоненты)

```python
df = pd.DataFrame(-1*lsa.U[:,1:3])
df = df.rename(columns={0:'x', 1:'y'})
df.insert(2,'dots', [lsa.fulldictionary[i] for i in lsa.keys])
df
```

### Координаты частей книги на плоскости

```python
doc_coord = pd.DataFrame(-1*lsa.Vt[1:3, :])
doc_coord = doc_coord.transpose()
doc_coord = doc_coord.rename(columns={0:'x', 1:'y'})
doc_coord.insert(2,'dots', [i for i in numpy.arange(1,len(text)+1)])
print(doc_coord)
```

# Расположение слов из книги на координатной плоскости

```python
source = ColumnDataSource(data=dict(x=df.x, y=df.y, dots=df.dots))
source2 = ColumnDataSource(data=dict(x=doc_coord.x, y=doc_coord.y, dots=doc_coord.dots))

hover = HoverTool(tooltips=[
    ("x","@x"),
    ("y", "@y"),
    ("dots", "@dots")
])

TOOLS = "pan, wheel_zoom, box_zoom, reset, save, box_select, crosshair"

output_notebook()
p = figure(plot_height=600, plot_width=800, title="Расположение слов на плоскости", tools=[hover, TOOLS])
p.circle(df.x, df.y, source=source, size=10, line_color="navy", fill_color="orange", fill_alpha=0.5)
show(p)
```

# Расположение глав из книги на координатной плоскости

```python
hover2 = HoverTool(tooltips=[
    ("x","@x"),
    ("y", "@y"),
    ("часть ", "@dots")
])
p4 = figure(plot_height=600, plot_width=800, title="Расположение частей книги на коорд. плоскости", tools=[hover2, TOOLS])
p4.circle(doc_coord.x, doc_coord.y, source=source2, line_color="green", size=30, fill_color="green", fill_alpha=0.3)
show(p4)
```

# Определение близости слова к каждой из частей книги

```python
word= "психология"
res = searchword(lsa, word)
p1 = figure(plot_height=600, plot_width=800, title="Расстояние от '"+word+"' до частей книги", tools="resize,save",x_axis_label='Расстояние', y_axis_label='Номер главы')
p1.segment(0, res[0], res[1], res[0], line_width=2, line_color="green", )
p1.circle(res[1], res[0], size=15, fill_color="orange", line_color="green", line_width=3, )

output_notebook()
show(p1)
```
